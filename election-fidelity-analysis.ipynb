```python
m_stack = itu_mobile.stack()
```

```python
def get_country_series(country):
    c1 = m_stack[country]
    c1.index = c1.index.map(lambda x:int(x))
    c1 = c1.map(lambda x: int(x.replace(',','')))
    return c1
    
def plot_country(country):
    s = get_country_series(country)
    s.plot()
```
