```python
import pandas as pd
import numpy as np

raw_df = pd.DataFrame.from_csv('31461-0001-Data.tsv',sep='\t', index_col=None)
```

```python
sa2 = raw_df[['newid','year','country','cow','r221r','r222r','r223r','r224r']]
sa2 = sa2[sa2.r221r != ' ']
sa2 = sa2[sa2.r222r != ' ']
sa2 = sa2[sa2.r223r != ' ']
sa2 = sa2[sa2.r224r != ' ']
```

```python
for c in raw_df.country.unique():
    c_df = trans_df[trans_df.country == c]
    r111r = c_df.r221r.max()
    r112r = c_df.r222r.max()
    r113r = c_df.r223r.max()
    r114r = c_df.r224r.max()
    print "{0}, {1}, {2}, {3}, {4}".format(c,r111r, r112r, r113r, r114r)
```
